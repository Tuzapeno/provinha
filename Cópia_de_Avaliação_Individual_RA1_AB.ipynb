# Questão 1: 
Um elemento radioativo tem uma meia-vida de 5 anos. Se inicialmente existem 500 gramas desse elemento, faça um gráfico que mostre como a quantidade desse elemento varia com o tempo e determine quantos gramas restarão após 15 anos. Utilize a equação $A = A0 * e^{-kt}$, onde $A0$ é a quantidade inicial do elemento, $k$ é a constante de decaimento (em 1/anos) e $t$ é o tempo decorrido em anos.



```python
import math
from sympy import Symbol, Limit
import matplotlib.pyplot as plt
import numpy as np

def fun(t):
  return 500 * (math.e ** (-0.5*t))

x_v = list(range(0, 17))
y_v = [fun(y) for y in x_v]

plt.plot(x_v, y_v)
plt.ylabel('Gramas')
plt.xlabel('Tempo')

plt.show()
```

# Questão 2
Usando, a seu critério as bibliotecas do Python, crie um gráfico para cada uma das funções a seguir. No label do gráfico deverá estar a função impressa com qualidade matemática (latex) e uma label dizendo se a função é crescente ou decrescente. 

1. $f(x)=2^x + 3^x + 25$

2. $f(x)=\log_2(x) - \log_2(3) - 1$

```python
import math
import matplotlib.pyplot as plt

def f1(x):
  return math.pow(2, x) + math.pow(3, x) + 25

def f2(x):
  if x == 0:
    return None
  else:
    return math.log(x, 2) - math.log(3, 2) - 1

x_vals = list(range(0, 100))
f1_y_vals = [f1(v) for v in x_vals]
f2_y_vals = [f2(v) for v in x_vals]

plt.plot(x_vals, f1_y_vals, '-', label=r"$f(x)=2^x + 3^x + 25$ ( Crescente )")
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()

plt.plot(x_vals, f2_y_vals, '-', c="green",label=r"$f(x)=\log_{2} x - \log_{2} 3 - 1$ ( Crescente )")
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

# Questão 3
Usando o Sympy, ou qualquer outro recurso disponível na internet, escreva, com qualidade matemática o passo-a-passo algébrico que leva a determinação dos seguintes limites. Ao final, para cada uma das equações você deverá traçar um gráfico da função. 

1. $\lim_{x \to 2} \frac{x^3 - 8}{x - 2}$

2. $\lim_{x \to \infty} \frac{3x^2 + 2x - 1}{x^2 - 4x + 7}$



\begin{equation}
\lim_{x \to 2} f(x) = \frac{x^3 - 8}{x - 2} = \frac{\left(x-2\right)\left(x^2+2x+4\right)}{x-2} = x^2 + 2x + 4 = 2^2 + 2(2) + 4 = 12
\end{equation}

\begin{equation}
\lim_{x \to ∞} f(x) = \frac{3x^2 + 2x - 1}{x^2 - 4x + 7} = \frac{\frac{3x^2}{x} + \frac{2x}{x} - \frac{-1}{x}}{\frac{x^2}{x} - \frac{4x}{x} + \frac{7}{x}} = \frac{3x + 2 - 0}{x - 4 - 0} = \frac{3x + 2}{x - 4} = \frac{\frac{3x}{x} + \frac{2}{x}}{\frac{x}{x} - \frac{4}{x}} = \frac{3 + 0}{1 - 0} =3
\end{equation}

```python
from sympy import Symbol, Limit
import matplotlib.pyplot as plt
import numpy as np

x = Symbol('x')

def limite_inf(f):
  x_v = np.linspace(-100, 100, 1000)
  y_v = [Limit(f, x, y).doit() for y in np.linspace(-100, 100, 1000)]

  plt.plot(x_v, y_v)
  plt.ylabel('f(x)')
  plt.xlabel('x')

  plt.show()

f = (x**3 - 8)/(x - 2)
limite_inf(f)
f2 = (3*x**2 + 2*x - 1)/(x**2 - 4*x + 7)
limite_inf(f2)
```
